```python
X = data[:, :-1]
y = data[:, -1]
```

```python
X = np.hstack((np.ones((y.size, 1)), X))
y = y.reshape(y.size, 1) # Turning the row vector y into a col vector
```

```python
m = y.size
```

```python
# Hypothesis Function
def h(z):
    return 1 / (1 + np.exp(-z))
```

```python
def cost_function(X, y, theta):
    # X - (m x n+1) matrix where each row contains the x values of one entry
    # y - (m x 1) matrix of correct outputs
    # theta - (n+1 x 1) matrix of theta values for each correponding x
    
    m = y.size
    return np.sum(-y*np.log(h(X@theta)) - (1-y)*np.log(1-h(X@theta))) / m
```

```python
def gradient_descent(X, y, theta, alpha, iterations):
    theta = theta.copy()
    m = y.size
    J_history = []
    
    for i in range(iterations):
        theta = theta - (alpha/m) * (X.T@(h(X@theta)-y))
        J_history.append(cost_function(X, y, theta))
    
    return theta, J_history
```

```python
def normalize_features(X):
    # X without the col of ones
    X_norm = X.copy()
    means = []
    sigmas = []
    for i in range(X_norm.shape[1]):
        means.append(np.mean(X_norm[:, i]))
        sigmas.append(np.std(X_norm[:, i]))
        X_norm[:, i] = (X_norm[:, i] - means[i]) / sigmas[i]
    return X_norm, means, sigmas
```

```python
# Normalizing X
X_norm, means, sigmas = normalize_features(X[:, 1:])
```

```python
X_norm = np.hstack((np.ones((m, 1)), X_norm))
```

```python
alpha = 10
iterations = 50
Itheta = np.zeros((X.shape[1], 1))

theta, J_history = gradient_descent(X_norm, y, Itheta, alpha, iterations)

print('Optimal theta:')
print(theta, '\n')
print(f'Cost at optimal theta: {J_history[-1]}')
```

```python
plt.plot(np.arange(len(J_history)), J_history, '-')
pass
```

```python
# Importing libraries
import os
import pandas as pd
import numpy as np
# ploting 2D data
from matplotlib import pyplot
# ploting 3D data
from mpl_toolkits.mplot3d import Axes3D
```

```python
df = pd.read_csv('Dataset.csv')
x = df['Fires per 1000 housing units'].to_numpy()
y = df['Thefts per 1000 population'].to_numpy()
m = y.size
```

```python
def visualize_data_by_point(x,y,c='red'):
    pyplot.scatter(x,y,c=c)
    
```

```python
def visualize_data_by_plot(x,y,c='red'):
    pyplot.plot(x, y, c=c)
    pyplot.xlabel('Number Of Itteration')
    pyplot.ylabel('Cost Function')
    pyplot.show()
```

```python
def add_theta_zero_column(x_array):
    return np.append(np.ones(m).reshape((m,1)),x.reshape((m,1)), axis=1)
x = add_theta_zero_column(x)
```

```python
def cost_function(x,y,theta, lambda_=0):
    
    j = 0
    m = y.size
    
    h = x @ theta
    differ = (h - y) ** 2
    j = np.sum(differ) / m
    j = j + (lambda_ / (2 * m)) * np.sum(theta[1:] ** 2)
    return j
```

```python
#testing cost
j = cost_function(x,y,np.array([0,0]), 0)
print(j)
```

```python
def gradient_decent(x,y,theta,alpha,num_iters):
    m = y.size
    tetha = theta.copy()
    info = []
    for i in range(num_iters):
        h = x @ theta
        difference = (h - y)
        theta = theta - (alpha / m) * (x.T @ difference)
        
        #then we compute cost function for each itteration to see the changes
        j = cost_function(x,y, theta)
        info.append([i, j])
        
    return theta, info
```

```python
# finding first y's thetas
theta, info = gradient_decent(x, y, np.array([0,0]) ,0.0001, 7000)
info = np.array([info])
info = info.T
visualize_data_by_plot(info[0], info[1])
print(theta)
```

```python
pred_x = np.linspace(0,40, m)
pred_x = np.append(np.ones((m,1)), pred_x.reshape((m,1)), axis=1)
```

```python
pred = np.dot(pred_x, theta)
```

```python
pyplot.figure()
pyplot.scatter(x[:, 1],y)
pyplot.plot(pred_x[:, 1], pred, c='red')
pyplot.legend(['Training data', 'Linear regression']);
pyplot.ylabel('Thefts per 1000 population')
pyplot.xlabel('Fires per 1000 housing units')
pyplot.show()
```
